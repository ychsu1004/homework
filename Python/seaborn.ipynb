```python
%matplotlib inline

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt



# 生成圖與子圖
fig = plt.figure()
axe = fig.add_subplot()

# 生成隨機資料
data = np.random.randn(100).cumsum()
print(data)

sns.set(style="white") # darkgrid, whitegrid, dark, white, ticks
# 以 seaborn 呈現
sns.histplot(data=data)
plt.show()
```

```python

# 生成隨機資料
data = np.random.randn(100).cumsum()

# 以 seaborn 呈現
sns.displot(data=data)
plt.show()
```
