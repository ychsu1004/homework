```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.array([4,5,9,12])
y = np.array([9,8,6,3])
```

```python
plt.scatter(x,y)
plt.plot(x, 11.803-0.707* x, 'g')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
X = x.reshape(4,1)
```

```python
regr.fit(X,y)
```

```python
Y = regr.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, Y, 'r')
plt.plot(x, 11.803-0.707* x, 'g')
```
